```python
from PIL import Image
import os

# Directory containing the images
image_folder = 'path/to/image/folder'

# Desired target size
target_size = (224, 224)

# Iterate over each image file in the folder
for filename in os.listdir(image_folder):
    # Construct the full path to the image file
    image_path = os.path.join(image_folder, filename)
    
    # Open the image using PIL
    with Image.open(image_path) as img:
        # Resize the image to the target size
        img_resized = img.resize(target_size)
        
        # Save the resized image, overwriting the original image file
        img_resized.save(image_path)
```

```python
# Directory containing the original images
original_directory = 'D:\\Desktop2\\kidneyDisease\\artifacts\\data_ingestion\\kidney-ct-scan-image'

# Directory to save the resized images
resized_directory = 'resized_images'

# Target size for resizing
target_size = (224, 224)

# Create the resized images directory if it doesn't exist
os.makedirs(resized_directory, exist_ok=True)


# Get list of subdirectories in the original directory
subdirectories = [d for d in os.listdir(original_directory) if os.path.isdir(os.path.join(original_directory, d))]

# Iterate over each subdirectory
for subdir in subdirectories:
    # Get list of image file names in the subdirectory
    image_filenames = os.listdir(os.path.join(original_directory, subdir))
    
    # Create subdirectory in the resized directory if it doesn't exist
    resized_subdir = os.path.join(resized_directory, subdir)
    os.makedirs(resized_subdir, exist_ok=True)
    
    # Iterate over each image file in the subdirectory
    for filename in image_filenames:
        
            # Load and resize the image
            img = load_img(os.path.join(original_directory, subdir, filename), target_size=target_size)
            
            # Save the resized image
            resized_image_path = os.path.join(resized_subdir, filename)
            save_img(resized_image_path, img)
            
            print(f"Resized image saved at: {resized_image_path}")
```
